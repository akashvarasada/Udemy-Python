```python
Age = 8

if Age < 15:
    print("Age less than 15")
elif Age < 10:
    print("Age less than 10")
elif Age < 16:
    print("Age less than 16")
elif Age < 17:
    print("Age less than 17")
elif Age < 18:
    print("Age less than 18")

else:
    print("Age greater than or equals to 15")
```

```python
# sequence of order is very important in elif confitions.
# It is similar to Switch function where first true condition will give output.
# here Age is 8 which is making true all the conditions but only first if will give output. rest conditions will not be checked.
# It is always better to add conditions in Ascending/desending order based on the requirements.
```

```python
#Example-1 determine the leap year using nested ifs

year = int(input("enter year: "))

if year%4 == 0:
    if year%100 != 0:
        print(f"{year} is a leap year")
    else:
        if year%400 == 0:
            print(f"{year} is a leap year")
        else:
            print(f"{year} is not a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
#Example-2 determine the leap year using single if statement.

year = int(input("enter year: "))

if year%4 == 0 and year%100 != 0 or year%400 == 0:
        print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year.")
```

```python
# Loop control statement
# Pass - No operation

for i in range(5):
    if i == 3:
        pass
        print("3 has come")
    else:
        print(i)
```

```python
# Loop control statement
# Continute - skips to next iteration
## anything written after continue will not execute.

for i in range(5):
    if i == 3:
        continue
        print("3 has come") #this line will not execute.
    else:
        print(i)
```

```python
# Loop control statement
# break - stops the loop.
## anything written after break will not execute.

for i in range(5):
    if i == 3:
        break
        print("3 has come") #this line will not execute.
    else:
        print(i)
```
